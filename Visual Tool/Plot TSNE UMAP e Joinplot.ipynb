```python
#ENVIRONMENT: cityhub    
# dash 2.14.2   ----> !pip install dash==2.14
#               ----> pip show dash

import pandas as pd
# only for test (Delete later)
dynamic_temp2 = pd.read_csv('s_d_UMAP_80.csv')
# options: 
# s_d_TSNE_50.csv
# s_d_UMAP_50.csv
# s_d_TSNE_80.csv
# s_d_UMAP_80.csv 

from dash import Dash, dcc, html, Input, Output, ctx, callback
import numpy as np
import pandas as pd
import plotly.express as px
import dash_bootstrap_components as dbc

# external_stylesheets = ["https://codepen.io/chriddyp/pen/bWLwgP.css"]

app = Dash(__name__)

# make a sample data frame with 6 columns
#np.random.seed(0)            # no-display
df = dynamic_temp2
dynamic_temp2 # = df_combinado #= dynamic_temp2.iloc[0:1000,:]
#df = df = px.data.iris()

app.layout = html.Div(
    [
        html.Div(
            dbc.Row(html.H5(children='Static Features'))),
        html.Div(
            dcc.Graph(id="g1", config={"displayModeBar": False}),
            className="four columns",
            ),
        html.Div(
            dbc.Row(html.H5(children='Dynamic Features (Crimes)'))),
        html.Div(  
            dcc.Graph(id="g2", config={"displayModeBar": False}),
            className="four columns",
        ),
    ],
    className="row",
)


def get_figure(df, x_col, y_col, selectedpoints, selectedpoints_local):
    #print('!!!!!!!!!!!!! selectedpoints_local ', selectedpoints_local)
    #print(' %%%%%%%%%%%% selectedpoints ', len(selectedpoints))

    if selectedpoints_local and "range" in selectedpoints_local:
        ranges = selectedpoints_local["range"]
        selection_bounds = {
            "x0": ranges["x"][0],
            "x1": ranges["x"][1],
            "y0": ranges["y"][0],
            "y1": ranges["y"][1],
        }
    else:
        selection_bounds = {
            "x0": np.min(df[x_col]),
            "x1": np.max(df[x_col]),
            "y0": np.min(df[y_col]),
            "y1": np.max(df[y_col]),
        }

    # set which points are selected with the `selectedpoints` property
    # and style those points with the `selected` and `unselected`
    # attribute. see
    # https://medium.com/@plotlygraphs/notes-from-the-latest-plotly-js-release-b035a5b43e21
    # for an explanation
    fig = px.scatter(df, x=df[x_col], y=df[y_col], text=df.index)#, color="species")

    fig.update_traces(
        selectedpoints=selectedpoints,
        customdata=df.index,
        mode="markers+text",
        # marker={"color": "rgba(0, 116, 217, 0.7)", "size": 20},
        marker={"size": 2,  "color":"red"},
        unselected={
            "marker": {"opacity": 0.1, "color":"blue"}, #0.3
            "textfont": {"color": "rgba(0, 0, 0, 0)"},
        },
    )

    fig.update_layout(
        margin={"l": 20, "r": 0, "b": 15, "t": 5},
        dragmode="select",
        hovermode=False,
        newselection_mode="gradual",
    )

    fig.add_shape(
        dict(
            {"type": "rect", "line": {"width": 1, "dash": "dot", "color": "red"}},
            **selection_bounds
        )
    )
    return fig


# this callback defines 3 figures
# as a function of the intersection of their 3 selections
@app.callback(
    [Output("g1", "figure"), Output("g2", "figure")],
    [Input("g1", "selectedData"), Input("g2", "selectedData")]
)
def callback(selection1, selection2):

    """if not ctx.triggered:
        raise dash.exceptions.PreventUpdate"""

    triggered_id = ctx.triggered[0]['prop_id'].split('.')[0]

    selectedpoints = df.index

    # Update selected points based on the triggered graph
    if triggered_id == 'g1' and selection1 and selection1["points"]:
        selectedpoints = np.intersect1d(selectedpoints,
                                       [p["customdata"] for p in selection1["points"]],
                                       assume_unique=True)
    elif triggered_id == 'g2' and selection2 and selection2["points"]:
        selectedpoints = np.intersect1d(selectedpoints,
                                       [p["customdata"] for p in selection2["points"]],
                                       assume_unique=True)

    return [get_figure(df, "x0", "y0", selectedpoints, selection1),
            get_figure(df, "x1", "y1", selectedpoints, selection2)]

if __name__ == "__main__":
    app.run(debug=True)
    #app.run_server()
```
